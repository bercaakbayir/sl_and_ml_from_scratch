# K-Means Algorithm from Scratch

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python

class KMeans:
    def __init__(self, k, max_iter=1000):
        self.k = k
        self.max_iter = max_iter
        self.centroids = None
        self.assignments = None
        
    def euclidean_distance(self, point1, point2):
        return np.linalg.norm(np.array(point1) - np.array(point2))
    
    def calculate_centroid(self, points):
        return np.mean(points, axis=0) if points else None  # Handle empty case

    def fit(self, data):
        n = len(data)
    
        # Initialize centroids randomly from data points
        self.centroids = random.sample(list(data), self.k)
    
        # Initialize assignments
        self.assignments = [random.randint(0, self.k - 1) for _ in range(n)]
    
        for _ in range(self.max_iter):
            new_centroids = []
        
            # Compute new centroids
            for i in range(self.k):
                cluster_points = [data[j] for j in range(n) if self.assignments[j] == i]
                if cluster_points:
                    new_centroids.append(self.calculate_centroid(cluster_points))
                else:
                    new_centroids.append(random.choice(data))  # Reinitialize to a random point
        
            # Assign points to closest centroid
            is_changed = False
            new_assignments = []
        
            for i in range(n):
                distances = [self.euclidean_distance(data[i], c) for c in new_centroids]
                closest_centroid = np.argmin(distances)
                new_assignments.append(closest_centroid)
            
                if self.assignments[i] != closest_centroid:
                    is_changed = True
        
            self.assignments = new_assignments
            self.centroids = new_centroids
        
            if not is_changed:  # Stop if no points changed clusters
                break

        return self.assignments, self.centroids  # ✅ Now returns values


    def predict(self, data):
        if self.centroids is None:
            raise ValueError("Model must be fitted before prediction.")
            
        predictions = []
        for point in data:
            distances = [self.euclidean_distance(point, c) for c in self.centroids]
            predictions.append(np.argmin(distances))
        return predictions
```

```python
def generate_data(num_clusters, points_per_cluster, spread):
    centers = np.random.rand(num_clusters, 2) * 10 # Küme merkezleri için rastgele değerler
    data = []
    for i in range(num_clusters):
        # Merkez etrafında dağılmış rastgele noktalar
        points = centers[i] + np.random.randn(points_per_cluster, 2) * spread
        data.extend(points)
    return np.array(data)


# Veri Üretimi
num_clusters = 4 # Küme sayısı
points_per_cluster = 50 # Her kümedeki nokta sayısı
spread = 1.5  # Dağılım parametresi
data = generate_data(num_clusters, points_per_cluster, spread)
```

```python

k = 3

kmeans = KMeans(k=k)
assignments, centroids = kmeans.fit(data)

print("Küme Atamaları:", assignments)
print("Merkez Noktaları:", centroids)
```

```python
new_data = np.array([[1.1, 2.3], [7, 9], [0.2,0.1]])
predictions = kmeans.predict(new_data)
print("Yeni veri noktaları için küme atamaları: ", predictions)
```

```python
def plot_clusters(data, assignments, centroids):
    data = np.array(data)
    centroids = np.array(centroids)

    plt.figure(figsize=(8, 6))

    # Scatter plot for each cluster
    for i in range(len(centroids)):
        cluster_points = data[np.array(assignments) == i]  # Get points in cluster
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i}')
    
    # Plot centroids
    plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', s=200, label='Centroids')

    plt.legend()
    plt.title("K-Means Clustering")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
plot_clusters(data, assignments, centroids)
```
