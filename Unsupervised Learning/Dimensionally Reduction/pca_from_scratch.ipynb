# Principal Component Analysis from Scratch

## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

## PCA Implementation 

```python
class PCA:
    
    def __init__(self, num_components):
        self.num_components = num_components
        
    def fit_transform(self, X):
        n_components = self.num_components

        
        X_mean = np.mean(X, axis=0)
        X_centered = X - X_mean
        
        mean = np.mean(X_centered, axis=0)  
        std_dev = np.std(X_centered, axis=0) 
        X_scaled = (X_centered - mean) / std_dev
        
        n = X.shape[0]
        cov_matrix =  (1/n) * np.dot(X_scaled.T, X_scaled)
        eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
        
        eigen_pairs = [(np.abs(eigenvalues[i]), eigenvectors[:, i]) for i in range(len(eigenvalues))]
        eigen_pairs.sort(key=lambda x: x[0], reverse=True)
        sorted_eigenvalues = np.array([pair[0] for pair in eigen_pairs])
        sorted_eigenvectors = np.array([pair[1] for pair in eigen_pairs]).T

        selected_eigenvalues = sorted_eigenvalues[:n_components]
        selected_eigenvectors = sorted_eigenvectors[:, :n_components]
        
        transformed_data = np.dot(X_scaled, selected_eigenvectors)
        
        pve = eigenvalues / np.sum(eigenvalues)
        

        return transformed_data, pve
```

```python
X = np.random.normal(loc = 0, scale = 1, size = (1000, 10))

pca = PCA(num_components=2)
X_transformed, pve = pca.fit_transform(X)

# Sonuçları yazdırma
print("Transformed Data:\n", X_transformed)

print("\nProportion Variance Explained:")
for i, variance in enumerate(pve):
    print(f"PC{i+1}: {variance:.4f} ({variance * 100:.2f}%)")
```
